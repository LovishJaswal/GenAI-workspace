# 📖 Tokenization — Complete Notes

## What is Tokenization?

Tokenization is the process of converting human-readable text into smaller units called **tokens**, which are then converted into **token IDs (integers)** that a language model can understand.

A Large Language Model (LLM) does **not** understand words or sentences directly. Neural networks only perform mathematical operations on numbers, so text must first be transformed into a numerical representation.

The complete pipeline is:

```
Raw Text
   ↓
Tokenizer(Encode)-----------------------------------------
   ↓
Tokens
   ↓
Token IDs (Integers)
   ↓
Embedding Layer
   ↓
Dense Vector Embeddings
   ↓
Transformer
   ↓
Next Token Prediction
   ↓
Tokenizer(Decode)------------------------------------------
   ↓
Human Readable Text
```

The tokenizer is therefore the **bridge between human language and neural networks**.

---

# Why Do We Need Tokenization?

Computers cannot perform calculations on text.

For example, the sentence:

```
"I love AI"
```

cannot be directly processed by a neural network.

Instead, it first becomes:

```
Tokens:
["I", " love", " AI"]

↓

Token IDs:
[40, 1842, 15592]
```

These Token IDs are then converted into vector embeddings by the embedding layer, and those vectors are what the Transformer actually processes.

Without tokenization, an LLM cannot read, understand, or generate language.

---

# Token vs Token ID

A **Token** is simply a piece of text.

Examples:

```
"I"

" love"

" AI"
```

A **Token ID** is the unique integer assigned to that token in the model's vocabulary.

Example:

```
"I"      → 40
" love"  → 1842
" AI"    → 15592
```

The model never processes the text directly—it processes the embeddings corresponding to these IDs.

---

# Ways to Split Text (Tokenization Strategies)

There are four major strategies for splitting text into tokens.

## 1. Character Tokenization

Every individual character becomes a token.

Example:

```
Hello

↓

H
e
l
l
o
```

Advantages:

* Very small vocabulary
* Works for every language

Disadvantages:

* Very long sequences
* Slow
* Memory inefficient

This approach is rarely used in modern LLMs.

---

## 2. Word Tokenization

Each word becomes one token.

Example:

```
I love AI

↓

["I", "love", "AI"]
```

Advantages:

* Simple
* Easy to understand

Disadvantages:

* Huge vocabulary
* Cannot handle unseen words
* Every variation ("play", "playing", "played") becomes a separate vocabulary entry

---

## 3. Subword Tokenization ⭐ (Most Common)

Words are broken into meaningful pieces.

Example:

```
playing

↓

play
ing
```

or

```
unbelievable

↓

un
believ
able
```

Advantages:

* Smaller vocabulary
* Handles unknown words
* Efficient
* Shorter sequences
* Shares pieces across similar words

Almost every modern LLM uses some form of Subword Tokenization.

---

## 4. Byte-Level Tokenization

Instead of splitting into words or characters, the tokenizer works with bytes.

Advantages:

* Supports every language
* Supports emojis
* Supports source code
* Supports special symbols

GPT-family models use **Byte-Level BPE**, which combines byte-level processing with the BPE algorithm.

---

# Tokenization Algorithms

The strategies above describe **how text is split**.

The following algorithms describe **how the tokenizer learns its vocabulary**.

## Byte Pair Encoding (BPE)

* Starts with characters.
* Repeatedly merges frequently occurring pairs.
* Builds a subword vocabulary automatically.

Used by:

* GPT-2
* GPT-3
* GPT-4
* GPT-5 (OpenAI encodings are based on byte-level BPE variants)

---

## WordPiece

Very similar to BPE, but chooses merges based on probability rather than frequency.

Used by:

* BERT
* DistilBERT

---

## Unigram

Starts with many candidate subwords and removes the least useful ones until an optimal vocabulary remains.

Used by:

* T5
* ALBERT

---

## SentencePiece

A tokenizer framework that does not rely on spaces between words, making it especially useful for multilingual languages such as Japanese and Chinese.

Many modern open-source models are trained using SentencePiece.

---

# Special Tokens

Most tokenizers also include predefined tokens used by the model.

Common examples include:

```
<BOS>   Beginning of Sequence

<EOS>   End of Sequence

<PAD>   Padding Token

<UNK>   Unknown Token

<MASK>  Mask Token (used by models like BERT)
```

Not every model uses every special token, but these are the most common ones.

# Why Can't We Use Any Tokenizer?

Every language model is trained using **one specific tokenizer**.

That tokenizer defines:

* Vocabulary
* Token IDs
* Merge rules
* Special tokens

The model's embedding layer is trained using those exact Token IDs.

For example:

```
Tokenizer A

"I love AI"

↓

[40, 1842, 15592]
```

A different tokenizer might produce:

```
Tokenizer B

"I love AI"

↓

[721, 84, 3001]
```

Even though both represent the same sentence, the IDs are different.

Since embeddings are indexed by Token IDs, changing the tokenizer means the model retrieves completely different embedding vectors.

**The tokenizer and the model must always match.**

---

# Why Do Modern Libraries Handle Tokenization Automatically?

Today, almost every AI framework performs tokenization internally.

Examples include:

* OpenAI API
* Hugging Face Transformers
* LangChain
* LlamaIndex
* Ollama

When you write something like:

```
model.generate("Hello")
```

the framework automatically:

1. Converts text into tokens.
2. Converts tokens into Token IDs.
3. Sends the IDs to the model.
4. Receives predicted Token IDs.
5. Decodes them back into readable text.

Most developers never need to manually tokenize text.

However, understanding tokenization is still essential because it directly affects:

* Context window limits
* API pricing (charged per token)
* RAG chunk sizes
* Prompt engineering
* Memory usage
* Latency
* Model compatibility
* Embeddings

# Key Takeaways

* Tokenization converts human-readable text into tokens.
* Tokens are mapped to unique Token IDs.
* The embedding layer converts Token IDs into dense vectors.
* The Transformer processes embeddings, **not raw text**.
* The four major tokenization strategies are Character, Word, Subword, and Byte-Level tokenization.
* The most common tokenization algorithms are BPE, WordPiece, Unigram, and SentencePiece.
* The three most common tokenizer implementations are **tiktoken (OpenAI)**, **AutoTokenizer (Hugging Face)**, and **SentencePiece**.
* Every model must use the tokenizer it was trained with because its embedding layer is built around that tokenizer's vocabulary and Token IDs.
* Modern frameworks perform tokenization automatically, but understanding it is fundamental for working with LLMs, embeddings, RAG systems, prompt engineering, context windows, and API costs.

##Method 2 - AutoTokenizer - Hugging Face models (Llama, Gemma, Mistral, Qwen, DeepSeek, BERT, etc.)

````python
from transformers import AutoTokenizer

# Load tokenizer for the model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "This is test"

# Split text into tokens
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

# Convert text into token IDs
token_ids = tokenizer.encode(text)
print("Token IDs:", token_ids)

# Count the number of tokens
print("Number of Tokens:", len(token_ids))

# Convert token IDs back to text
decoded = tokenizer.decode(token_ids)
print("Decoded Text:", decoded)

# Convert tokens into token IDs
ids = tokenizer.convert_tokens_to_ids(tokens)
print("IDs from Tokens:", ids)
````

##Method 3 - SentencePiece - Direct use of SentencePiece tokenizers or custom-trained tokenizers

Unlike tiktoken and AutoTokenizer, SentencePiece requires a trained tokenizer model (tokenizer.model). It cannot be used directly unless you already have that file.It is a tokenizer library developed by Google. Many models such as T5, mT5, ALBERT, Llama, Gemma, and Mistral use SentencePiece-based tokenizers internally. In practice, developers rarely use the sentencepiece library directly. Instead, they use Hugging Face's AutoTokenizer, which automatically downloads and loads the appropriate SentencePiece tokenizer when required.
